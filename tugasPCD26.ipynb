```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

membaca dua buah gambar

```python
img1 = cv2.imread("sample_data/motor.png")
img2 = cv2.imread("sample_data/mobil.png")
resize_img1 = cv2.resize(img1,(300,300))
resize_img2 = cv2.resize(img1,(300,300))
resize_img3 = cv2.resize(img2,(300,300))


# membedakan gambar
diff_img = cv2.absdiff(resize_img1,resize_img2)
diff_img1 = cv2.absdiff(resize_img1,resize_img3)
cv2_imshow(diff_img)


```

filtering rat-rata

```python
average_img = cv2.blur(resize_img1, (5, 5))
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1), plt.imshow(cv2.cvtColor(resize_img1, cv2.COLOR_BGR2RGB)), plt.title('Original Image'), plt.axis('off')
plt.subplot(1, 2, 2), plt.imshow(cv2.cvtColor(average_img, cv2.COLOR_BGR2RGB)), plt.title('Median Blur Image'), plt.axis('off')
plt.show()
```

filter media

```python
median_img = cv2.medianBlur(resize_img1, 5)
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1), plt.imshow(cv2.cvtColor(resize_img1, cv2.COLOR_BGR2RGB)), plt.title('Original Image'), plt.axis('off')
plt.subplot(1, 2, 2), plt.imshow(cv2.cvtColor(median_img, cv2.COLOR_BGR2RGB)), plt.title('Median Blur Image'), plt.axis('off')
plt.show()
```

Filter minimum

```python
min_img = cv2.erode(resize_img1, np.ones((5,5),np.uint64))
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1), plt.imshow(cv2.cvtColor(resize_img1, cv2.COLOR_BGR2RGB)), plt.title('Original Image'), plt.axis('off')
plt.subplot(1, 2, 2), plt.imshow(cv2.cvtColor(min_img, cv2.COLOR_BGR2RGB)), plt.title('filter minimum Image'), plt.axis('off')
plt.show()
```

filtering maksimum

```python
max_img = cv2.dilate(resize_img1, np.ones((5,5),np.uint8))
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1), plt.imshow(cv2.cvtColor(resize_img1, cv2.COLOR_BGR2RGB)), plt.title('Original Image'), plt.axis('off')
plt.subplot(1, 2, 2), plt.imshow(cv2.cvtColor(max_img, cv2.COLOR_BGR2RGB)), plt.title('filter maksimum Image'), plt.axis('off')
plt.show()
```

Operasi Gradien Sobel

```python
gray_img = cv2.cvtColor(resize_img1, cv2.COLOR_BGR2GRAY)
grad_x = cv2.Sobel(gray_img, cv2.CV_64F, 1, 0, ksize=3)
grad_y = cv2.Sobel(gray_img, cv2.CV_64F, 0, 1, ksize=3)
gradient_img = cv2.sqrt(grad_x**2 + grad_y**2)
arr_img = [gradient_img,resize_img1]
for i in range(2):
  cv2_imshow(arr_img[i])



```

```python
laplacian_img = cv2.Laplacian(resize_img1, cv2.CV_64F)
arr_img = [laplacian_img,resize_img1]
for i in range(2):
  cv2_imshow(arr_img[i])







```
